```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
frequent_itemsets = fpgrowth(basket, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets.shape
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

```python
import time
from mlxtend.frequent_patterns import apriori

start_time_fp = time.time()
fp_itemsets = fpgrowth(basket, min_support=0.1, use_colnames=True)
fp_duration = time.time() - start_time_fp

start_time_ap = time.time()
ap_itemsets = apriori(basket, min_support=0.1, use_colnames=True, low_memory=True)
ap_duration = time.time() - start_time_ap

print(f"Время работы FP-Growth: {fp_duration:.4f} секунд")
print(f"Время работы Apriori: {ap_duration:.4f} секунд")
```

```python
start_time_fp = time.time()
fp_itemsets = fpgrowth(basket, min_support=0.03, use_colnames=True)
fp_duration = time.time() - start_time_fp

start_time_ap = time.time()
ap_itemsets = apriori(basket, min_support=0.03, use_colnames=True, low_memory=True)
ap_duration = time.time() - start_time_ap

print(f"FP-Growth (support=0.02): {fp_duration:.4f} секунд")
print(f"Apriori (support=0.02): {ap_duration:.4f} секунд")
```
